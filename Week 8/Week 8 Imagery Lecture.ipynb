# Week 8 Imagery Lecture

```python
from arcgis.gis import GIS
from arcgis.raster import ImageryLayer
#from arcgis.geocoding import Geocoder, get_geocoders

gis = GIS("http://slustl.maps.arcgis.com/home", "gregbrunner_slugis")
```

```python
il = ImageryLayer(url='http://naip.arcgis.com/arcgis/rest/services/NAIP/ImageServer', gis=gis)
```

```python
the_map = gis.map("St. Louis, MO")
the_map
```

```python
the_map.add_layer(il)
```

```python
the_map.extent
```

```python
values = il.get_samples(geometry=the_map.extent, sample_count=1000)
```

```python
values
```

```python
val_list = []
for val in values:
    val_list.append(val['values'])
```
